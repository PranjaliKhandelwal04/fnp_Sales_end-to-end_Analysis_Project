```python
#Importing the Libraries:
import pandas as pd
import numpy as np
```

```python
#importing and reading the csv files:
customers = pd.read_csv("customers.csv")
orders = pd.read_csv("orders.csv")
products = pd.read_csv("products.csv")
```

```python
#printing the first 5 rows of the files:
print(customers.head(5))
print(orders.head(5))
print(products.head(5))
```

```python
#merging all the three files in df:
df = orders.merge(customers,on="Customer_ID",how = "left")
df = df.merge(products,on="Product_ID",how = "left")
df.head(5)
```

```python
#Statistical Analysis of the dataset:
df.describe()
```

```python
#Finding the total number of null values in df:
df.isnull().sum()
```

```python
#Finding the total number of duplicate values in df:
df.duplicated().sum()
```

```python
#Converting the Order_Date column to datetime datatype:
df['Order_Date'] = pd.to_datetime(df['Order_Date'],errors='coerce')
```

```python
#Renaming the column:
df.rename(columns={'Price (INR)':'UnitPrice'},inplace=True)
df.columns
```

```python
#Checking for invalid values:
if 'Quantity' in df.columns:
  df = df[df['Quantity']>0]
```

```python
#Checking for invalid values:
if 'UnitPrice' in df.columns:
  df = df[df['UnitPrice']>0]
```

```python
#Creating a new column of total amount:
if {'Quantity','UnitPrice'}.issubset(df.columns):
  df['TotalAmount'] = df['Quantity']*df['UnitPrice']
```

```python
#Standardize the text columns:
for col in df.select_dtypes(include='object').columns:
  df[col] = df[col].astype(str).str.strip()
  df[col] = df[col].replace('nan','Unknown')
```

```python
#Creating new columns:
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
df['Month_Name'] = df['Order_Date'].dt.month_name()
df['Day'] = df['Order_Date'].dt.date
df['Weekday_Name'] = df['Order_Date'].dt.day_name()
```

```python
#Exporting cleaned dataset:
df.to_csv("fnp_cleaned.csv",index=False)
```

```python
#Calculating Total Revenue:
total_revenue = df['TotalAmount'].sum()
```

```python
#Calculating total orders:
total_orders = df['Order_ID'].nunique()
```

```python
#Calculating total customers:
total_customers = df['Customer_ID'].nunique()
```

```python
#Calculating Average order value:
avg_order_value = total_revenue/total_orders
```

```python
#Top 10 products by revenue:
top_products = (df.groupby('Product_ID')['TotalAmount']
                  .sum()
                  .sort_values(ascending=False)
                  .head(10))
```

```python
#Calculating Sales by Category:
sales_by_category = df.groupby('Category')['TotalAmount'].sum().sort_values(ascending=False)
```

```python
#Calculating monthly trends:
monthly_sales = (df.groupby(['Year','Month_Name'])['TotalAmount']
                   .sum()
                   .reset_index())
```

```python
#printing all the values:
print("Total Revenue:",total_revenue)
print("Total Orders:",total_orders)
print("Total Customers:",total_customers)
print("Average Order Value:",avg_order_value)
print("\nTop 10 Products:")
print(top_products)
print("\nSales by Category")
print(sales_by_category)
print("\nMonthly Sales Trend:")
print(monthly_sales)
```
